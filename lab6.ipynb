```python
import boto3
from botocore.exceptions import ClientError
import json

def get_secret(secret_name):
    region_name = "us-east-1"

    # Create a Secrets Manager client
    session = boto3.session.Session()
    client = session.client(
        service_name='secretsmanager',
        region_name=region_name
    )

    try:
        get_secret_value_response = client.get_secret_value(
            SecretId=secret_name
        )
    except ClientError as e:
        raise e

    secret = get_secret_value_response['SecretString']
    
    return json.loads(secret)
```

```python
import psycopg2

conn = psycopg2.connect(host=get_secret('postgresql')['host'],
                      user=get_secret('postgresql')['username'],
                      password = get_secret('postgresql')['password'],
                      dbname=get_secret('postgresql')['engine'])

cur = conn.cursor()
```

```python
sql = " select * from name inner join income on name.fips = income.fips "
```

```python
import pandas as pd

df = pd.read_sql(sql,conn)

df.head()
```

```python
%%ai gpt4 -f code

calculate the income growth rate in all states from the {df} in {In[13]}
display the result on a chart
```

```python
import matplotlib.pyplot as plt

# Calculate growth rate
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Filter for the first and last years
df_filtered = df[df['year'].isin([2005, 2019])]

# Pivot the DataFrame to get the income growth for each state
pivot_df = df_filtered.pivot(index='name', columns='year', values='income_growth')

# Plot
plt.figure(figsize=(10,8))
pivot_df.loc[:,2019].sort_values().plot(kind='barh', color='skyblue')
plt.xlabel('Income Growth Rate')
plt.title('Income Growth Rate from 2005 to 2019 by State')
plt.show()
```

```python
%%ai gpt4 -f code

calculate the year-over-year income growht rate for each state of each year in dataframe {df} in {In[13]}
use a line chart to show the result
```

```python
import matplotlib.pyplot as plt

# Calculate year-over-year income growth rate for each state
df['income_growth'] = df.groupby('name')['income'].pct_change()

# Plot
plt.figure(figsize=(12,8))
for state in df['name'].unique():
    plt.plot(df[df['name'] == state]['year'], df[df['name'] == state]['income_growth'], label=state)
plt.xlabel('Year')
plt.ylabel('Income Growth Rate')
plt.title('Year-Over-Year Income Growth Rate by State')
plt.legend(loc='best', bbox_to_anchor=(1, 1))
plt.show()
```
